```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Interatomic potential

```python
e = 2
s = 2
r = np.linspace(0.5, 10, 1000)
E_rep = 4 * e * (s / r) ** 12
E_att = - 4 * e * (s / r) ** 6
V = E_rep + E_att

plt.rcParams.update({'figure.figsize':(8,6), 'font.size':20})
fig, ax = plt.subplots()
ax.plot(r, V, lw=8)
ax.plot([2.5, 3], [2, 2], color='gray', marker='o', ms=60, mfc='#ED4E33', mew=0.1)
ax.hlines(y=0, xmin=0, xmax=10, color='gray', ls='dotted', lw=2)
ax.set(xlim=[1.5, 4], ylim=[-4,5], xticks=[], yticks=[])
plt.show()
fig.savefig('C:/Users/enzec/Desktop/iap.svg')
```
